```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#load in dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/elleobrien/wine/master/wine_quality.csv")
dataset
```

```python
#one hot encode dataset
one_hot_encoded_dataset = pd.get_dummies(dataset)
one_hot_encoded_dataset
```

```python
#separate into train and test

#seperate into x(input), and y(predicted value/label)
x = one_hot_encoded_dataset.drop('quality', axis = 1)
y = one_hot_encoded_dataset['quality']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
x_train.shape
```

```python
from keras import layers, losses
wine_quality_model = keras.models.Sequential()
wine_quality_model.add(layers.Flatten(input_shape = (11,)))
wine_quality_model.add(layers.BatchNormalization())
wine_quality_model.add(layers.Dense(256, activation = 'relu'))
wine_quality_model.add(layers.Dense(256, activation = 'relu'))
wine_quality_model.add(layers.Dropout(0.5))
wine_quality_model.add(layers.Dense(256, activation = 'relu'))
wine_quality_model.add(layers.Dropout(0.5))
wine_quality_model.add(layers.Dense(128, activation = 'relu'))
wine_quality_model.add(layers.Dense(10, activation = 'softmax'))

wine_quality_model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
wine_quality_model.summary()
```

```python
wine_quality_model.fit(x_train, y_train, epochs = 15, validation_data= (x_test, y_test))
```

```python
wine_quality_model.evaluate(x_test, y_test)
```

```python
import math
def turn_list_into_percentages(list):
  newlist = []

  for i in list:
    i = round(float(i), 2)
    newlist.append(i)

  jeffrey = []
  for i in newlist:
    i = i*100
    jeffrey.append(i)

  lol = []
  for i in jeffrey:
    i = math.floor(i)
    lol.append(i)
  

  return lol
```

```python
#load in y train
y_train
```

```python
import random
q = random.randint(1, 1279)
b = q+1


prediction = wine_quality_model.predict(x_train[q:b])


llist = turn_list_into_percentages(list(prediction[0]))

biggest_value = 0
for i in llist:
  if i > biggest_value:
    biggest_value = i

wine_rating = llist.index(biggest_value) +1

print(llist)
wine_rating_true = y_train[q]
print(f"i predict that wine {q} has a {biggest_value} percent chance of being rated {wine_rating} out of 10")
print(f"wine {q} has a rating of {wine_rating_true}")


if wine_rating != wine_rating_true:
  llist.remove(biggest_value)

  biggest_value = 0
  for i in llist:
    if i > biggest_value:
      biggest_value = i

  wine_rating = llist.index(biggest_value)

  print(f"sorry, I can only predict with 65% accuracy, wine {q} also has a {biggest_value} percent chance of being rated {wine_rating} out of 10")
```
